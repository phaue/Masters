The errros used are errors from the calibration, a difference in method used, and scaled errors from the gaussian fit from the chisquared analysis.

```python
def weighted_mean(values, errors):
    values = np.asarray(values)
    errors = np.asarray(errors)
    weights = 1.0 / (errors**2)
    mean = np.sum(weights * values) / np.sum(weights)
    err_mean = np.sqrt(1.0 / np.sum(weights))
    return mean, err_mean

means = [1.167,1.169,1.167]; means_err = [0.003, 0.004, 0.003]
trues = [8.443, 8.445, 8.443]; trues_err = [0.003,0.004,0.003]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [1.250,1.251,1.249]; means_err = [0.002, 0.002, 0.002]
trues = [7.146, 7.147, 7.145]; trues_err = [0.002,0.002,0.002]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [1.313,1.316,1.311]; means_err = [0.002, 0.002, 0.003]
trues = [6.880, 6.883, 6.878]; trues_err = [0.003,0.002,0.003]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()


means = [1.850,1.850,1.850]; means_err = [0.003, 0.002, 0.002]
trues = [7.775, 7.775, 7.774]; trues_err = [0.003,0.002,0.002]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

print(f"Weighted mean peak value(only id3):     {2.486}({0.002*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {8.442}({0.002*1e3+0.1:.0f})")

print("="*60)
print()

print(f"Weighted mean peak value(only id3):     {2.396}({0.003*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {8.348}({0.003*1e3+0.1:.0f})")

print("="*60)
print()

print(f"Weighted mean peak value(only id3):     {2.563}({0.004*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {9.906}({0.004*1e3+0.1:.0f})")

print("="*60)
print()
print(f"Weighted mean peak value(only id3):     {2.610:.3f}({0.002*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {8.240:.3f}({0.002*1e3+0.1:.0f})")

print("="*60)
print()

means = [3.877,3.870,3.880]; means_err = [0.013, 0.006, 0.005]
trues = [9.899, 9.892, 9.902]; trues_err = [0.013,0.006,0.005]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [5.464,5.466,5.456]; means_err = [0.012, 0.006, 0.002]
trues = [14.060, 14.061, 14.051]; trues_err = [0.013,0.006,0.003]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [1.655, 1.667, 1.673]; means_err = [0.005, 0.004, 0.005]
trues = [7.571, 7.583, 7.590]; trues_err = [0.005,0.005,0.005]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()
means = [6.526,6.522,6.521]; means_err = [0.014, 0.006,0.006]
trues = [14.059,14.055,14.054]; trues_err = [0.015,0.006,0.006]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()
means = [7.847,7.845,7.842]; means_err = [0.012, 0.006, 0.006]
trues = [14.060, 14.057, 14.055]; trues_err = [0.013,0.007,0.006]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()
```

```python
######### Peak values read from id3 spectrum very roughly
import numpy as np
frac = 0.9542011788469517 # (Na21/(p+Na21)) amount of energy the daughter nucleus gets of the total energy released in the decay
frac_err = 1.032071836027589e-10 # uncertainty of the fraction
psep = 5.50410 # proton separation energy of 22Mg
psep_err = 0.00016 # error of proton separation energy of 22Mg

readEs = [1.425,1.490,2.950,3.320,3.375,4.360,4.450,4.475,5.545,5.6]
readerrs = [0.01,0.015,0.01,0.015,0.01,0.015,0.015,0.01,0.01,0.01]
gammas = [4,3,2,1,1,1,3,1,2,0]

def trueE(E,Eerr, gamma=0):
    addon = 0
    if gamma==1:
        addon = 332
    if gamma==2:
        addon = 332+1384
    if gamma==3:
        addon = 332+1384+1113
    if gamma==4:
        addon = 332+1384+1113+1590
    if gamma>4:
        return "aint no gamma that big"
    TE = E/frac+psep+addon/1e3
    TE_err = np.sqrt((1/frac * Eerr)**2 + (E/(frac**2) * frac_err)**2 + (1/frac * psep_err)**2)

    return TE, TE_err, gamma

for i,j,g in zip(readEs, readerrs,gammas):
    print(trueE(i,j,g))
```
