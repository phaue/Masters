```python
def weighted_mean(values, errors, inflate=True, returnchi=False):
    values = np.asarray(values)
    errors = np.asarray(errors)
    weights = 1.0 / (errors**2)

    mean = np.sum(weights * values) / np.sum(weights)
    err_mean = np.sqrt(1.0 / np.sum(weights))

    # chi2
    chi2 = np.sum(((values - mean)/errors)**2)
    dof = len(values) - 1
    chi2_red = chi2 / dof if dof > 0 else np.nan

    # inflate if chi2_red > 1
    if inflate and chi2_red > 1:
        err_mean *= np.sqrt(chi2_red)
    if returnchi:
        return mean, err_mean, chi2, dof, chi2_red
    else:
        return mean, err_mean
means = [1.167,1.168,1.167]; means_err = [0.003, 0.003, 0.003]
trues = [8.443, 8.445, 8.443]; trues_err = [0.003,0.004,0.003]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [1.250,1.251,1.249]; means_err = [0.002, 0.002, 0.002]
trues = [7.146, 7.147, 7.145]; trues_err = [0.002,0.002,0.002]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [1.313,1.316,1.311]; means_err = [0.002, 0.002, 0.002]
trues = [6.880, 6.883, 6.878]; trues_err = [0.003,0.002,0.002]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()


means = [1.851,1.850,1.850]; means_err = [0.003, 0.003, 0.002]
trues = [7.776, 7.775, 7.774]; trues_err = [0.003,0.003,0.003]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

print(f"Weighted mean peak value(only id3):     {2.486}({0.003*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {8.442}({0.003*1e3+0.1:.0f})")

print("="*60)
print()

print(f"Weighted mean peak value(only id3):     {2.397}({0.003*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {8.348}({0.003*1e3+0.1:.0f})")

print("="*60)
print()

print(f"Weighted mean peak value(only id3):     {2.563}({0.003*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {9.906}({0.003*1e3+0.1:.0f})")

print("="*60)
print()
print(f"Weighted mean peak value(only id3):     {2.610:.3f}({0.002*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {8.240:.3f}({0.002*1e3+0.1:.0f})")

print("="*60)
print()

means = [3.877,3.870,3.880]; means_err = [0.012, 0.006, 0.004]
trues = [9.899, 9.892, 9.902]; trues_err = [0.013,0.006,0.004]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [5.464,5.457,5.456]; means_err = [0.012, 0.008, 0.002]
trues = [14.060, 14.052, 14.051]; trues_err = [0.013,0.009,0.003]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [1.656, 1.668, 1.675]; means_err = [0.004, 0.008, 0.0012]
trues = [7.572, 7.584, 7.590]; trues_err = [0.005,0.008,0.0013]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()
means = [6.524,6.522,6.517]; means_err = [0.013, 0.006,0.007]
trues = [14.057,14.055,14.051]; trues_err = [0.014,0.006,0.007]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()
means = [7.847,7.844,7.843]; means_err = [0.012, 0.006, 0.006]
trues = [14.060, 14.057, 14.055]; trues_err = [0.013,0.007,0.006]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [8.165,8.154]; means_err = [0.013, 0.009] ## id1&id3
trues = [14.060, 14.049]; trues_err = [0.012,0.009]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()
```

```python
######### Peak values read from id3 spectrum very roughly
import numpy as np
frac = 0.9542011788469517 # (Na21/(p+Na21)) amount of energy the daughter nucleus gets of the total energy released in the decay
frac_err = 1.032071836027589e-10 # uncertainty of the fraction
psep = 5.50410 # proton separation energy of 22Mg
psep_err = 0.00016 # error of proton separation energy of 22Mg
trues = np.array([1425, 1500, 2970, 3055, 3315, 3380,3450, 4350,4425,4470, 5555, 5630, 7200, 7260])/1e3
trues_err = np.array([20, 25, 30, 30, 30, 30, 20,25, 20, 20, 25, 25, 25, 25])/1e3

gammas = [4,3,2,2,1,1,2,1,3,1,2,2,0,0]

def trueE(E,Eerr, gamma=0):
    addon = 0
    if gamma==1:
        addon = 332
    if gamma==2:
        addon = 332+1384
    if gamma==3:
        addon = 332+1384+1113
    if gamma==4:
        addon = 332+1384+1113+1590
    if gamma>4:
        return "aint no gamma that big"
    TE = E/frac+psep+addon/1e3
    TE_err = np.sqrt((1/frac * Eerr)**2 + (E/(frac**2) * frac_err)**2 + (1/frac * psep_err)**2)

    return TE, TE_err, gamma

for i,j,g in zip(trues, trues_err,gammas):
    ts = trueE(i,j,g)
    print(f"Read proton E: {i:.3f} -- Deduced Ex22Mg: {ts[0]:.3f}({ts[1]*1e3:.0f}), {g}")
```

```python
vals = np.array([0.452, 0.692,0.770,0.931,0.995,1.167,1.250,1.313,1.673,1.850,2.396,2.486,2.563,3.877,5.456,6.520,7.844,8.158])
val_errs = np.array([2,1,2,2,1,2,1,2,4,2,3,3,3,3,2,4,4,7])/1e3
peaks = [6.229,6.310,6.879,7.146,7.589,7.775,8.348,8.442,9.309,9.902,14.054]
peak_errs = np.array([1,2,1,1,3,2,3,2,2,4,2])/1e3
frac = 0.9542011788469517
frac_err = 1.032071836027589e-10
vals /=frac
z_err = np.sqrt((val_errs/frac)**2 + (vals*frac_err/frac**2)**2)

navals = [(peaks[0]-vals[1]),(peaks[1]-vals[0],peaks[1]-vals[2]), (peaks[2]-vals[4],peaks[2]-vals[7]),(peaks[3]-vals[6]),(peaks[4]-vals[8]),
          (peaks[5]-vals[9]),(peaks[6]-vals[10]),(peaks[7]-vals[5],peaks[7]-vals[11]),(peaks[8]-vals[3]),(peaks[9]-vals[12],peaks[9]-vals[13]),
          (peaks[10]-vals[14],peaks[10]-vals[15],peaks[10]-vals[16],peaks[10]-vals[17])]
navals_errs = [
    np.sqrt(peak_errs[0]**2 + z_err[1]**2),
    (np.sqrt(peak_errs[1]**2 + z_err[0]**2), np.sqrt(peak_errs[1]**2 + z_err[2]**2)),
    (np.sqrt(peak_errs[2]**2 + z_err[4]**2), np.sqrt(peak_errs[2]**2 + z_err[7]**2)),
    np.sqrt(peak_errs[3]**2 + z_err[6]**2),
    np.sqrt(peak_errs[4]**2 + z_err[8]**2),
    np.sqrt(peak_errs[5]**2 + z_err[9]**2),
    np.sqrt(peak_errs[6]**2 + z_err[10]**2),
    (np.sqrt(peak_errs[7]**2 + z_err[5]**2), np.sqrt(peak_errs[7]**2 + z_err[11]**2)),
    np.sqrt(peak_errs[8]**2 + z_err[3]**2),
    (np.sqrt(peak_errs[9]**2 + z_err[12]**2), np.sqrt(peak_errs[9]**2 + z_err[13]**2)),
    (np.sqrt(peak_errs[10]**2 + z_err[14]**2), np.sqrt(peak_errs[10]**2 + z_err[15]**2),
     np.sqrt(peak_errs[10]**2 + z_err[16]**2), np.sqrt(peak_errs[10]**2 + z_err[17]**2))
]
for n,p,ne in zip(navals,peaks,navals_errs):
    print(f"Deduced excited state energy in 22Mg:       {p}")
    
    print(f"Deduced 21Na (excited) state energies: ")
    if isinstance(n, tuple):
        for na,nes in zip(n,ne):
            print(f"                                            {na:.3f}({nes*1e3:.0f})")
    else:
        print(f"                                            {n:.3f}({ne*1e3:.0f})")
    print()
    print("--"*30)
```
