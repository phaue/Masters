```python
def weighted_mean(values, errors):
    values = np.asarray(values)
    errors = np.asarray(errors)
    weights = 1.0 / (errors**2)
    mean = np.sum(weights * values) / np.sum(weights)
    err_mean = np.sqrt(1.0 / np.sum(weights))
    return mean, err_mean

means = [1.166,1.167,1.168]; means_err = [0.003, 0.002, 0.003]
trues = [8.442, 8.444, 8.444]; trues_err = [0.003,0.002,0.003]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [1.250,1.251,1.250]; means_err = [0.002, 0.002, 0.002]
trues = [7.146, 7.147, 7.146]; trues_err = [0.002,0.002,0.002]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [1.314,1.315,1.312]; means_err = [0.003, 0.002, 0.002]
trues = [6.881, 6.883, 6.879]; trues_err = [0.003,0.002,0.002]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()


means = [1.850,1.850,1.850]; means_err = [0.002, 0.002, 0.002]
trues = [7.774, 7.775, 7.774]; trues_err = [0.002,0.002,0.002]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

print(f"Weighted mean peak value(only id3):     {2.486}({0.002*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {8.441}({0.002*1e3+0.1:.0f})")

print("="*60)
print()

print(f"Weighted mean peak value(only id3):     {2.386}({0.003*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {8.337}({0.003*1e3+0.1:.0f})")

print("="*60)
print()

print(f"Weighted mean peak value(only id3):     {2.565}({0.003*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {9.908}({0.003*1e3+0.1:.0f})")

print("="*60)
print()
print(f"Weighted mean peak value(only id3):     {2.610:.3f}({0.002*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value(only id3):  {8.240:.3f}({0.002*1e3+0.1:.0f})")

print("="*60)
print()

means = [3.877,3.881,3.870]; means_err = [0.006, 0.006, 0.003]
trues = [9.899, 9.903, 9.892]; trues_err = [0.006,0.006,0.004]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

means = [5.464,5.466,5.456]; means_err = [0.006, 0.006, 0.002]
trues = [9.899, 9.903, 9.892]; trues_err = [0.006,0.007,0.003]
wm = weighted_mean(means,means_err)
tm = weighted_mean(trues,trues_err)
print(f"Weighted mean peak value:               {wm[0]:.3f}({wm[1]*1e3+0.1:.0f})")
print(f"Weighted excited 22Mg value:            {tm[0]:.3f}({tm[1]*1e3+0.1:.0f})")

print("="*60)
print()

```
